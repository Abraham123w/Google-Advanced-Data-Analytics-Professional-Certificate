```python
#==> ENTER YOUR CODE HERE

# How many of each payment type are represented in the data?

# Contar cuántos viajes hay por cada tipo de pago
print(df['payment_type'].value_counts())
```

According to the data dictionary, the payment method was encoded as follows:

1 = Credit card

4 = Dispute

```python
#==> ENTER YOUR CODE HERE

# What is the average tip for trips paid for with credit card?

# Filtrar solo los viajes pagados con tarjeta de crédito (payment_type == 1)
# y luego calcular el promedio (mean) de la columna 'tip_amount'
avg_cc_tip = df[df['payment_type'] == 1]['tip_amount'].mean()

print('Propina promedio con tarjeta de crédito:', avg_cc_tip)

#==> ENTER YOUR CODE HERE

# What is the average tip for trips paid for with cash?



# Filtrar solo los viajes pagados con efectivo (payment_type == 2)
# y calcular el promedio de la columna 'tip_amount'
avg_cash_tip = df[df['payment_type'] == 2]['tip_amount'].mean()

print('Propina promedio con efectivo:', avg_cash_tip)
```

```python
#==> ENTER YOUR CODE HERE

# How many times is each vendor ID represented in the data?

# Contar los viajes por cada VendorID
print(df['VendorID'].value_counts())
```

```python
#==> ENTER YOUR CODE HERE

# What is the mean total amount for each vendor?

# Calcular el promedio de total_amount para cada VendorID
print(df.groupby('VendorID')['total_amount'].mean())
```
